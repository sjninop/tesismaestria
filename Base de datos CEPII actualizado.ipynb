```python
def sumar_pesos(lista_de_dataframes):
    # Inicializa un DataFrame vacío para almacenar los resultados
    df_total_pesos = pd.DataFrame()

    # Itera sobre la lista de DataFrames y aplica la función de agregación
    for df in lista_de_dataframes:
        # Verifica si la columna "Value" es numérica
        if pd.api.types.is_numeric_dtype(df['Value']):
            # Convierte la columna "Value" a tipo numérico (int o float)
            df['Value'] = pd.to_numeric(df['Value'])
            # Agrupa los datos por país de origen y país de destino y suma los pesos
            df_i_total_pesos = df.groupby(['Exporter', 'Importer'])['Value'].sum().reset_index()

            # Agrega el resultado al DataFrame vacío
            df_total_pesos = pd.concat([df_total_pesos, df_i_total_pesos], ignore_index=True)
        else:
            print("Advertencia: La columna 'Value' no es numérica en el siguiente DataFrame:")
            print(df)

    # Devuelve el resultado
    return df_total_pesos
```
